```python
#importing standard libraries
import pandas as pd
import numpy as np
import xarray as xr 
import matplotlib.pyplot as plt
%matplotlib inline
```

### _open cmip6 and gauge dataset || select the precipitation variables from both datasets || resample both data variables by month_

```python
# cmip6 data
cmip6_data = xr.open_dataset('./cmip6.nc')
pr = cmip6_data['pr']
cmip6_precip = pr.groupby('time.month').mean()

# guage data
guage_data = xr.open_dataset('./gauge_gh.nc', engine='netcdf4')
gp = guage_data['precip']
gauge_precip = gp.groupby('time.month').mean()
```

### _extract Saltpond from both CMIP6 and GUAGE precipitation_

```python
# extract saltpond
saltpond_cmip = cmip6_precip.sel(lon=-1.05, lat=5.20, method='nearest')
saltpond_gauge = gauge_precip.sel(longitude=-1.05, latitude=5.20, method='nearest')

# get the values for both cmip6 and guage precip
saltpond_cmip_values = saltpond_cmip.data
saltpond_gauge_values = saltpond_gauge.values

# convert these values to dataframe
month = ['Jan', 'Feb', 'March', 'April', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
data = {'month': month, 
        'saltpond-cmip': saltpond_cmip_values, 
        'saltpond-gauge': saltpond_gauge_values}
saltpond_df = pd.DataFrame(data=data)
saltpond_df
```

### _Algorithm to calculate for the root mean square error (RMSE)_

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```

```python
predictions = saltpond_df['saltpond-cmip']
targets = saltpond_df['saltpond-gauge']

error = rmse(predictions, targets)
print("RMSE:", error)
```

### _Algorithm to calculate for the bias_

```python
def calculate_bias(predictions, targets):
    return predictions.mean() - targets.mean()
```

```python
predictions = saltpond_df['saltpond-cmip']
targets = saltpond_df['saltpond-gauge']

bias = calculate_bias(predictions, targets)
print("Bias:", bias)
```
